# Introduction to grid_py

This notebook introduces the core building blocks of the **grid_py** package,
the Python port of R's *grid* graphics system. You will learn about:

* **Units** -- the measurement system that underpins all positioning and sizing.
* **Viewports** -- rectangular sub-regions of a figure that define their own coordinate systems.
* **Graphical parameters (Gpar)** -- a structured way to specify colours, line widths, fonts, etc.
* **Drawing primitives** -- functions such as `grid_rect` and `grid_text` that place graphical objects on the page.

The narrative follows the original R *grid* vignette (`grid.Rnw`).

```python
import matplotlib
matplotlib.use("Agg")

import grid_py as gp
```

## Units

Everything in *grid_py* is sized and positioned using `Unit` objects.
A `Unit` stores a numeric value together with a *unit type* such as
`"cm"`, `"inches"`, `"npc"` (Normalised Parent Coordinates), or `"native"`.

Units support arithmetic: you can add, subtract, multiply, and negate them.

```python
# Create some basic units
u1 = gp.Unit(1, "inches")
u2 = gp.Unit(2, "cm")
u3 = gp.Unit(0.5, "npc")

print("u1:", u1)
print("u2:", u2)
print("u3:", u3)
```

```python
# Unit arithmetic
u_sum = u1 + u2
u_neg = -u1
u_scaled = 3 * u2

print("u1 + u2:", u_sum)
print("-u1:    ", u_neg)
print("3 * u2: ", u_scaled)
```

```python
# Combining units into a vector with unit_c
u_vec = gp.unit_c(gp.Unit(1, "cm"), gp.Unit(0.5, "npc"), gp.Unit(2, "inches"))
print("unit_c result:", u_vec)
print("length:       ", gp.unit_length(u_vec))
```

```python
# Parallel min / max / sum
a = gp.Unit(2, "cm")
b = gp.Unit(1, "inches")
print("pmax:", gp.unit_pmax(a, b))
print("pmin:", gp.unit_pmin(a, b))
print("psum:", gp.unit_psum(a, b))
```

## Viewports

A `Viewport` defines a rectangular sub-region of the current figure.
It carries its own coordinate system (`xscale`, `yscale`), clipping,
and graphical parameter settings. Viewports are *pushed* onto a stack
to become active and *popped* when no longer needed.

```python
# Create a viewport that occupies 80% of the parent region
vp = gp.Viewport(
    width=gp.Unit(0.8, "npc"),
    height=gp.Unit(0.8, "npc"),
    name="plot_region",
)
print(vp)
```

```python
# A viewport with custom native scales
vp_native = gp.Viewport(
    xscale=[0, 100],
    yscale=[0, 50],
    name="native_region",
)
print(vp_native)
```

## Graphical Parameters (Gpar)

`Gpar` bundles graphical settings such as colours, line types, font sizes,
and transparency. Parameters can be scalar or vectorised (one value per
graphical object).

```python
# Create graphical parameter sets
gp_red = gp.Gpar(col="red", lwd=2)
gp_fancy = gp.Gpar(col="blue", fill="lightblue", fontsize=14, lty="dashed")

print("gp_red:  ", gp_red)
print("gp_fancy:", gp_fancy)
```

## Drawing Primitives

With units, viewports, and graphical parameters in hand, you can draw
on the page. `grid_newpage()` clears the figure and resets the viewport
stack. Then functions like `grid_rect()` and `grid_text()` add graphical
objects.

```python
# Start a new page and draw a rectangle with text inside it
gp.grid_newpage()

# Draw a rectangle filling the entire page
gp.grid_rect(gp=gp.Gpar(fill="grey90"))

# Push a smaller viewport and draw inside it
gp.push_viewport(gp.Viewport(
    width=gp.Unit(0.6, "npc"),
    height=gp.Unit(0.6, "npc"),
    name="inner",
))

gp.grid_rect(gp=gp.Gpar(col="steelblue", fill="aliceblue", lwd=2))
gp.grid_text(
    "Hello, grid_py!",
    gp=gp.Gpar(fontsize=18, col="steelblue"),
)

gp.pop_viewport()
```

```python
# Demonstrate positioning with units
gp.grid_newpage()

# Three rectangles at different positions
for i, (xpos, colour) in enumerate([
    (0.2, "firebrick"),
    (0.5, "forestgreen"),
    (0.8, "royalblue"),
]):
    gp.grid_rect(
        x=gp.Unit(xpos, "npc"),
        y=gp.Unit(0.5, "npc"),
        width=gp.Unit(0.2, "npc"),
        height=gp.Unit(0.4, "npc"),
        gp=gp.Gpar(col=colour, fill=colour, alpha=0.4),
    )
    gp.grid_text(
        str(i + 1),
        x=gp.Unit(xpos, "npc"),
        y=gp.Unit(0.5, "npc"),
        gp=gp.Gpar(fontsize=20, col="white"),
    )
```

3. **Gpar** -- a structured set of graphical parameters.
