# Interactive Grid: Editing Grobs

This tutorial demonstrates how to interactively edit graphical objects (grobs)
in `grid_py`, following the narrative of the R `grid` vignette *Editing grid Graphics*.

Key concepts covered:
- Creating named grobs so they can be referenced later
- Editing graphical parameters (colour, rotation) with `edit_grob`
- Editing children of a composite grob via `GPath`

```python
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

import grid_py as gp
```

## Step 1 -- Create an x-axis grob

We create an x-axis with tick marks at 0.2, 0.4, 0.6, and 0.8 and assign
it the name `"gxa"` so we can refer to it in subsequent edits.

```python
gxa = gp.xaxis_grob(
    at=[0.2, 0.4, 0.6, 0.8],
    vp=gp.Viewport(width=0.5, height=0.01),
    name="gxa",
)

gp.grid_newpage()
gp.grid_draw(gxa)
plt.gcf()
```

## Step 2 -- Edit the axis colour to red

Using `edit_grob` we change the colour of the entire axis to red.
Notice we refer to the grob object directly in this off-screen workflow.

```python
gxa = gp.edit_grob(gxa, gp=gp.Gpar(col="red"))

gp.grid_newpage()
gp.grid_draw(gxa)
plt.gcf()
```

## Step 3 -- Edit the labels to green

We use `GPath` to target only the `"labels"` child of the axis grob,
changing its colour to green while the rest stays red.

```python
gxa = gp.edit_grob(gxa, gPath="labels", gp=gp.Gpar(col="green"))

gp.grid_newpage()
gp.grid_draw(gxa)
plt.gcf()
```

## Step 4 -- Change the tick positions

We can also change structural properties of the axis.  Here we set new
tick positions at 0.0, 0.5, and 1.0.  The labels revert to the axis
colour (red) because new label children are created, inheriting from
the parent.

```python
gxa = gp.edit_grob(gxa, at=[0.0, 0.5, 1.0])

gp.grid_newpage()
gp.grid_draw(gxa)
plt.gcf()
```

## Step 5 -- Set labels to black and rotate 30 degrees

Finally we target the labels again, setting their colour to black and
rotating them by 30 degrees.

```python
gxa = gp.edit_grob(gxa, gPath="labels", gp=gp.Gpar(col="black"), rot=30)

gp.grid_newpage()
gp.grid_draw(gxa)
plt.gcf()
```

## Complete off-screen workflow

The entire editing sequence can be written as a single pipeline,
building the grob object entirely off-screen before drawing once.

```python
gxa = gp.xaxis_grob(
    at=[0.2, 0.4, 0.6, 0.8],
    vp=gp.Viewport(width=0.5, height=0.01),
    name="gxa",
)
gxa = gp.edit_grob(gxa, gp=gp.Gpar(col="red"))
gxa = gp.edit_grob(gxa, gPath="labels", gp=gp.Gpar(col="green"))
gxa = gp.edit_grob(gxa, at=[0.0, 0.5, 1.0])
gxa = gp.edit_grob(gxa, gPath="labels", gp=gp.Gpar(col="black"), rot=30)

gp.grid_newpage()
gp.grid_draw(gxa)
plt.gcf()
```
