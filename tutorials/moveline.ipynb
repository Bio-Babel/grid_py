# Move-To and Line-To

This tutorial demonstrates `grid_move_to` and `grid_line_to`, the
`grid_py` equivalents of R's `grid.move.to` and `grid.line.to`.

These primitives maintain a *current drawing location* that persists
across viewports, enabling lines to be drawn between different
coordinate systems.

```python
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

import grid_py as gp
```

## Drawing a line across two viewports

We create a layout with three columns: two "null" columns for panels
and a 1-inch gap in between.  We place a point in the left panel and
call `grid_move_to` to record that position.  Then we navigate to the
right panel (which has a different y-scale) and call `grid_line_to` to
draw a line from the remembered position to the new point.

```python
gp.grid_newpage()

# Outer viewport with a 1x3 layout
lay = gp.GridLayout(
    nrow=1, ncol=3,
    widths=gp.Unit([1, 1, 1], ["null", "inches", "null"]),
)
gp.push_viewport(
    gp.Viewport(width=0.8, height=0.8, layout=lay)
)

# --- Left panel ---
gp.push_viewport(
    gp.Viewport(layout_pos_col=1, yscale=[0, 4])
)
gp.grid_grill()
gp.grid_yaxis()
gp.grid_xaxis()
gp.grid_points(0.5, gp.Unit(2, "native"))
gp.grid_move_to(0.5, gp.Unit(2, "native"))
gp.pop_viewport()

# --- Right panel ---
gp.push_viewport(
    gp.Viewport(layout_pos_col=3, yscale=[0, 2])
)
gp.grid_grill()
gp.grid_yaxis()
gp.grid_xaxis()
gp.grid_points(0.5, gp.Unit(2, "native"))
gp.grid_line_to(0.5, gp.Unit(2, "native"))
gp.pop_viewport()

plt.gcf()
```

The line crosses from the left panel (y-scale 0--4) to the right panel
(y-scale 0--2), passing through the gap column.  The current drawing
location set by `grid_move_to` persists even after we pop and push
different viewports.

## Using move-to / line-to within a single viewport

The same primitives work within a single viewport for incremental
line drawing.

```python
gp.grid_newpage()

gp.push_viewport(gp.Viewport(width=0.6, height=0.6))
gp.grid_rect()  # border

# Draw a zigzag path using move_to / line_to
xs = [0.1, 0.3, 0.5, 0.7, 0.9]
ys = [0.2, 0.8, 0.2, 0.8, 0.2]

gp.grid_move_to(xs[0], ys[0])
for x, y in zip(xs[1:], ys[1:]):
    gp.grid_line_to(x, y, gp=gp.Gpar(col="steelblue", lwd=2))

gp.pop_viewport()
plt.gcf()
```
